```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
url="https://www.anfp.cl/estadisticas"
```

```python
# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text
```

```python
# Parse the html content
soup = BeautifulSoup(html_content, "html.parser")
print(soup.prettify()) # print the parsed data of html
```

```python
print(soup.title)
```

```python
for link in soup.find_all("a"):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
```

```python
table = soup.findAll("table", attrs={"class": "table tbls table-striped table-bordered dt-responsive display nowrap"})[0]
table
```

```python
values =[]

tr = table.findAll(['tr'])
tr
```

```python
csvFile = open("futbolchileno_v0.csv",'wt',newline='',encoding='latin1')
writer = csv.writer(csvFile)  
try:   
        for cell in tr:
            th = cell.find_all('th')
            th_data = [col.text.strip('\n') for col in th]
            td = cell.find_all('td')
            row = [i.text.replace('\n','') for i in td]
            writer.writerow(th_data+row)      
        
finally:   
    csvFile.close()
```
